```python
import networkx as nx
from matplotlib import pyplot as plt
```

```python
node_type = {
    1: {"color": "w", "stroke": "black"},
    2: {"color": "yellow", "stroke": "black"},
}
edge_type = {1: {"width": 1.0, "color": "black"}, 2: {"width": 2.0, "color": "red"}}
```

```python
# node_type={1: {'color': 'w', 'stroke': 'black'}, 2: {'color': 'r', 'stroke': 'black'}}
# inputs
G = nx.Graph()
v = [i for i in range(7)]
e = [
    (0, 1, 1),
    (2, 1, 1),
    (2, 3, 1),
    (4, 5, 2),
    (3, 0, 1),
    (1, 5, 5),
    (6, 1, 1),
    (6, 4, 1),
    (3, 5, 1),
    (6, 2, 1),
]
G.add_nodes_from(v, **node_type[1])
G.add_weighted_edges_from(e, **edge_type[1])

# change attributes
G.add_node(1, **node_type[2])
G.add_edge(2, 6, **edge_type[2])

# draw
pos=nx.circular_layout(G)
draw_kwargs = {
    "G": G,
    "pos": pos,
    "with_labels": True,
    "node_color": list(nx.get_node_attributes(G, "color").values()),
    "edgecolors": list(nx.get_node_attributes(G, "stroke").values()),
    "width": list(nx.get_edge_attributes(G, "width").values()),
    "edge_color": list(nx.get_edge_attributes(G, "color").values()),
    # "edge_labels": nx.get_edge_attributes(G, "weight")
}
nx.draw(**draw_kwargs)

edge_labels=nx.get_edge_attributes(G, "weight")
nx.draw_networkx_edge_labels(G,pos,edge_labels)
plt.savefig("./1.png")
```
