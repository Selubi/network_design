
## 演習８ 問題３(k辺連結保存))

```python
from numpy import argmax
from copy import deepcopy

# inputs
G = nx.Graph()
v = [i for i in range(7)]
e = [
    (0, 1, 1),
    (2, 1, 1),
    (2, 3, 1),
    (4, 5, 1),
    (3, 0, 1),
    (1, 5, 1),
    (6, 1, 1),
    (6, 4, 1),
    (3, 5, 1),
    (6, 2, 1),
    (0, 4, 1),
]
G.add_nodes_from(v, **node_type[1])
G.add_weighted_edges_from(e, **edge_type[1])


pos = nx.circular_layout(G)


# MA alg
adj_list = []
o_count = 1


while len(adj_list) != len(G):
    if not adj_list:
        node = 0
        adj_list.append(node)
        G.add_node(node, order=o_count)
        adj_edge = []
    else:
        adj = []
        for i in G:
            if i in adj_list:
                adj.append([])
            else:
                adj.append([e for e in G[i] if e in adj_list])
        node = argmax(adj)
        adj_list.append(node)
        G.add_node(node, order=o_count)
        adj_edge = [(node, i) for i in adj[node]]

    # draw steps
    G_draw = deepcopy(G)
    G_draw.add_node(node, **node_type[2])
    for e in adj_edge:
        G_draw.add_edge(*e, **edge_type[3])
    draw_kwargs = {
        "pos": pos,
        "with_labels": True,
        "labels": nx.get_node_attributes(G_draw, "order"),
        "node_color": list(nx.get_node_attributes(G_draw, "color").values()),
        "edgecolors": list(nx.get_node_attributes(G_draw, "stroke").values()),
        "width": list(nx.get_edge_attributes(G_draw, "width").values()),
        "edge_color": list(nx.get_edge_attributes(G_draw, "color").values()),
        "style": list(nx.get_edge_attributes(G_draw, "style").values()),
    }

    nx.draw(G_draw, **draw_kwargs)
    plt.savefig(f"./ex-8-NA-{o_count}.png")

    o_count += 1

# reorder based on MA algorithm
priority = nx.get_node_attributes(G_draw, "order")
inverse_priority = {priority[i]: i for i in priority}
neworder = sorted(priority, key=priority.get)

# print(neworder)
index = 1
for i in reversed(neworder):
    # i is the original index
    adj = [priority[e] for e in G[i] if priority[e] < priority[i]]
    adj = sorted(adj)
    # j is priority index
    count = 1
    for j in adj:
        G.add_edge(i, inverse_priority[j], order=count)
        count += 1

    # draw steps
    G_draw = deepcopy(G)
    adj_edge = [(i, inverse_priority[j]) for j in adj]
    for e in adj_edge:
        G_draw.add_edge(*e, **edge_type[3])
    draw_kwargs = {
        "pos": pos,
        "with_labels": True,
        "labels": nx.get_node_attributes(G_draw, "order"),
        "node_color": list(nx.get_node_attributes(G_draw, "color").values()),
        "edgecolors": list(nx.get_node_attributes(G_draw, "stroke").values()),
        "width": list(nx.get_edge_attributes(G_draw, "width").values()),
        "edge_color": list(nx.get_edge_attributes(G_draw, "color").values()),
        "style": list(nx.get_edge_attributes(G_draw, "style").values()),
    }
    edge_labels = nx.get_edge_attributes(G, "order")
    nx.draw(G_draw, **draw_kwargs)
    nx.draw_networkx_edge_labels(G, pos, edge_labels)
    plt.savefig(f"./ex-8-F-{index}.png")

    index += 1


plt.clf()


edges = nx.get_edge_attributes(G, "order")
edges_F = [e for e in edges.keys() if edges[e] <= 1]
# for e in edges_F:
#     G.add_edge(*e,**edge_type[3])
draw_kwargs = {
    "pos": pos,
    "with_labels": True,
    "edgelist": edges_F,
    "labels": nx.get_node_attributes(G, "order"),
    "node_color": list(nx.get_node_attributes(G, "color").values()),
    "edgecolors": list(nx.get_node_attributes(G, "stroke").values()),
    "width": list(nx.get_edge_attributes(G, "width").values()),
    "edge_color": list(nx.get_edge_attributes(G, "color").values()),
    "style": list(nx.get_edge_attributes(G, "style").values()),
}
nx.draw(G, **draw_kwargs)
print(edges_F)
plt.savefig(f"./ex-8-1S.png")

edges_F = [e for e in edges.keys() if edges[e] <= 2]
# for e in edges_F:
#     G.add_edge(*e,**edge_type[3])
draw_kwargs = {
    "pos": pos,
    "with_labels": True,
    "edgelist": edges_F,
    "labels": nx.get_node_attributes(G, "order"),
    "node_color": list(nx.get_node_attributes(G, "color").values()),
    "edgecolors": list(nx.get_node_attributes(G, "stroke").values()),
    "width": list(nx.get_edge_attributes(G, "width").values()),
    "edge_color": list(nx.get_edge_attributes(G, "color").values()),
    "style": list(nx.get_edge_attributes(G, "style").values()),
}
nx.draw(G, **draw_kwargs)
plt.savefig(f"./ex-8-2S.png")
# draw
# draw
# shape of the graph


# # edge labels
# edge_labels = nx.get_edge_attributes(G, "weight")
# # nx.draw_networkx_edge_labels(G,pos,edge_labels)
# plt.savefig("./ex8-1.png")
```

## 演習９問題１(辺連結分解)

```python
from copy import deepcopy

# inputs
G = nx.Graph()
v = [i for i in range(7)]
e = [
    (0, 1, 1),
    (2, 1, 1),
    (2, 3, 1),
    (4, 5, 1),
    (3, 0, 1),
    (1, 5, 1),
    (6, 1, 1),
    (6, 4, 1),
    (3, 5, 1),
    (6, 2, 1),
]
G.add_nodes_from(v, **node_type[2])
G.add_weighted_edges_from(e, **edge_type[1])


# calculate minimum cut for every pair.
min_cut = {
    pairs: nx.minimum_cut(G, *pairs, capacity="weight")[0]
    for pairs in [(i, j) for i in G for j in G if i != j]
}

# k辺連結性分解(c-辺のグループ)
def k_edge_decomposition(G, k):
    nodes = [i for i in G]
    searched = []
    ans = []  # stores the final answer group
    reset = True
    while len(searched) != len(nodes):
        # makes a new group and initiate searching parameters
        if reset:
            search_range = [i for i in nodes if i not in searched]
            now_searching = [search_range[0]]
            searched.append(search_range[0])
            reset = False
        tmp_searched = searched[:]  # to compare if a new element is added to the group

        # for every unsearched element
        for i in [i for i in search_range if i not in now_searching]:
            add_flag = True
            # if there is a cut below c between i and any of now_searching, its not c-edge-connected with the current group
            for j in now_searching:
                if min_cut[(i, j)] < k:
                    add_flag = False
                    break
            # greedy, if we find a matching vertex that we can add, add that. breaks out from the loop
            if add_flag:
                searched.append(i)
                now_searching.append(i)
                break
        # if no vertex is inserted into group, this is a group. start searching from another vertex
        if len(tmp_searched) == len(searched):
            ans.append(now_searching)
            reset = True
        # if a vertex is inserted continue the search (or if all vertex is exhausted add the last group to ans)
        else:
            if len(searched) == len(nodes):
                ans.append(now_searching)
            reset = False
    return ans


# draw
pos = nx.planar_layout(G)
group_id = 2
G_draw = deepcopy(G)
for i in k_edge_decomposition(G, 3):
    for j in i:
        G_draw.add_node(j, **node_type[group_id])
    group_id += 1
draw_kwargs = {
    "pos": pos,
    "with_labels": True,
    "node_color": list(nx.get_node_attributes(G_draw, "color").values()),
    "edgecolors": list(nx.get_node_attributes(G_draw, "stroke").values()),
    "width": list(nx.get_edge_attributes(G_draw, "width").values()),
    "edge_color": list(nx.get_edge_attributes(G_draw, "color").values()),
    "style": list(nx.get_edge_attributes(G_draw, "style").values()),
}
nx.draw(G_draw, **draw_kwargs)
plt.savefig(f"./ex-9-K-{3}.png")


nx.draw(**draw_kwargs_gen(G, pos))

# edge labels
edge_labels = nx.get_edge_attributes(G, "weight")
# nx.draw_networkx_edge_labels(G,pos,edge_labels)
plt.savefig("./ex-9-K-2.png")
```

## 演習１１問題６(最小コストフロー)

```python
from copy import deepcopy

arc_rad = 0.1


def draw_kwargs_gen(gr, pos):
    return {
        "G": gr,
        "pos": pos,
        "with_labels": True,
        "node_color": list(nx.get_node_attributes(gr, "color").values()),
        "edgecolors": list(nx.get_node_attributes(gr, "stroke").values()),
        "width": list(nx.get_edge_attributes(gr, "width").values()),
        "edge_color": list(nx.get_edge_attributes(gr, "color").values()),
        "connectionstyle": f"arc3, rad = {arc_rad}",
        # "style": list(nx.get_edge_attributes(G, "style").values()),
    }


def BFS(G, src, sink):
    adj_list = []
    for n, nbrsdict in G.adjacency():
        adj_list.append({nbr: eattr["weight"] for nbr, eattr in nbrsdict.items()})
    search_queue = [i for i in adj_list[src].keys()]
    searched = [src] + search_queue
    parent = [src if i in search_queue else -1 for i in G.nodes()]
    flow = [adj_list[src][i] if i in search_queue else -1 for i in G.nodes()]
    while search_queue:
        now = search_queue.pop(0)
        new_found = [i for i in adj_list[now].keys() if i not in searched]
        search_queue += new_found
        searched += new_found
        for i in new_found:
            parent[i] = now
            flow[i] = min(flow[now], adj_list[now][i])
    if flow[sink] == -1:
        return False
    path = [sink]
    now = sink
    while parent[now] != -1:
        path.insert(0, parent[now])
        now = parent[now]
    return path, flow[sink]


def draw_residual_graph(gr):
    pos = nx.shell_layout(gr)
    nx.draw(**draw_kwargs_gen(gr, pos))
    edge_weight = nx.get_edge_attributes(gr, "weight")
    edge_cost = nx.get_edge_attributes(gr, "cost")
    edge_labels = {i: f"({edge_weight[i]},{edge_cost[i]})" for i in edge_weight.keys()}
    nx.draw_networkx_edge_labels(gr, pos, edge_labels=edge_labels, label_pos=0.25)


def draw_MF_graph(gr):
    pos = nx.shell_layout(gr)
    nx.draw(**draw_kwargs_gen(gr, pos))
    edge_weight = nx.get_edge_attributes(gr, "weight")
    edge_lim = nx.get_edge_attributes(gr, "lim")
    edge_labels = {i: f"{edge_weight[i]} ({edge_lim[i]})" for i in edge_weight.keys()}
    nx.draw_networkx_edge_labels(gr, pos, edge_labels=edge_labels, label_pos=0.25)


def draw_copy(gr, edgelist, index):
    G_draw = deepcopy(gr)
    for i in edgelist:
        G_draw.add_edge(*i, **edge_type[3])
    draw_residual_graph(G_draw)
    plt.savefig(f"./ex12-MF-{index}")
    plt.clf()


def maximum_flow(G, G_f, source, sink):
    while BFS(G, source, sink):
        path, flow_n = BFS(G, source, sink)
        path_flow = pathtotuple(path)
        path_residue = pathtotuple(path[::-1])
        # print(path_flow)
        # draw_copy(G,path_flow,counter)
        # counter+=1
        # adds residue path to residue graph
        for i in path_residue:
            if G.has_edge(*i):
                G.add_edge(*i, weight=flow_n + G.get_edge_data(*i)["weight"])
            else:
                G.add_edge(
                    *i,
                    weight=flow_n,
                    cost=-(G.get_edge_data(i[1], i[0])["cost"]),
                    residue=True,
                    **edge_type[1],
                )
        for i in path_flow:
            # adds flow to the max flow graph
            if G.get_edge_data(*i)["residue"]:
                G_f.add_edge(*i, weight=G_f.get_edge_data(*i)["weight"] - flow_n)
            else:
                G_f.add_edge(*i, weight=G_f.get_edge_data(*i)["weight"] + flow_n)
            # deletes flow from residue graph
            if G.get_edge_data(*i)["weight"] == flow_n:
                G.remove_edge(*i)
            else:
                G.add_edge(*i, weight=G.get_edge_data(*i)["weight"] - flow_n)


def Bellman_Ford(G, src):
    v = G.nodes()
    e = G.edges()
    dist = [0 if i == src else float("inf") for i in v]
    parent = [-1 for i in v]
    # flag=False
    for i in range(len(v) - 1):
        for u, v in e:
            if dist[v] > dist[u] + G.get_edge_data(u, v)["cost"]:
                dist[v] = dist[u] + G.get_edge_data(u, v)["cost"]
                parent[v] = u
    # for u,v in e:
    #     if(dist[v]>dist[u]+G.get_edge_data(u,v)["cost"]):
    #         flag=True
    if dist[src] < 0:
        now = src
        path = [src]
        while parent[now] != src:
            path.insert(0, parent[now])
            now = parent[now]
        path.insert(0, src)
        # print(path)
        return path
    return []


def get_max_flow(gr, path):
    path_tuple = pathtotuple(path)
    flow_n = float("inf")
    for u, v in path_tuple:
        flow_n = gr[u][v]["weight"] if gr[u][v]["weight"] < flow_n else flow_n
    return flow_n


G = nx.DiGraph()
v = [i for i in range(5)]
e = [
    (0, 1, {"weight": 15, "cost": 4}),
    (0, 2, {"weight": 8, "cost": 4}),
    (1, 2, {"weight": 20, "cost": 2}),
    (1, 3, {"weight": 4, "cost": 2}),
    (1, 4, {"weight": 10, "cost": 6}),
    (2, 3, {"weight": 15, "cost": 1}),
    (3, 4, {"weight": 20, "cost": 2}),
    (4, 2, {"weight": 5, "cost": 3}),
    (2, 4, {"weight": 4, "cost": 3}),
]


G.add_nodes_from(v, **node_type[1])
for i in e:
    G.add_edge(i[0], i[1], **i[2], residue=False, **edge_type[1])
# G.add_weighted_edges_from(e, **edge_type[1])


G_f = nx.DiGraph()
G_f.add_nodes_from(v, **node_type[1])
for i in e:
    G_f.add_edge(
        i[0], i[1], weight=0, lim=G.get_edge_data(i[0], i[1])["weight"], **edge_type[1]
    )


source = 0
sink = 4
G.add_node(source, **node_type[2])
G.add_node(sink, **node_type[3])
G_f.add_node(source, **node_type[2])
G_f.add_node(sink, **node_type[3])

# draw base graph
draw_residual_graph(G)
plt.savefig(f"./ex12-MF-start")
plt.clf()

# find maximum flow
maximum_flow(G, G_f, source, sink)
draw_MF_graph(G_f)
plt.savefig(f"./ex12-MF-flow-start")
plt.clf()

# find minimum cost for maximum flow
index = 1
while Bellman_Ford(G, sink):
    path = Bellman_Ford(G, sink)
    flow_n = get_max_flow(G, path)
    path_flow = pathtotuple(path)
    path_residue = pathtotuple(path[::-1])
    draw_copy(G, path_flow, index)
    index += 1
    for i in path_residue:
        if G.has_edge(*i):
            G.add_edge(*i, weight=flow_n + G.get_edge_data(*i)["weight"])
        else:
            G.add_edge(
                *i,
                weight=flow_n,
                cost=-(G.get_edge_data(i[1], i[0])["cost"]),
                residue=True,
                **edge_type[1],
            )
    for i in path_flow:
        # adds flow to the max flow graph
        if G_f.has_edge(*i):
            G_f.add_edge(*i, weight=G_f.get_edge_data(*i)["weight"] + flow_n)
        else:
            G_f.add_edge(
                i[1], i[0], weight=G_f.get_edge_data(i[1], i[0])["weight"] - flow_n
            )
        if G.get_edge_data(*i)["weight"] == flow_n:
            G.remove_edge(*i)
        else:
            G.add_edge(*i, weight=G.get_edge_data(*i)["weight"] - flow_n)


rmv_lst = []
for i in G_f.edges:
    if G_f.get_edge_data(*i)["weight"] == 0:
        rmv_lst.append(i)
for i in rmv_lst:
    G_f.remove_edge(*i)
draw_residual_graph(G)
plt.savefig(f"./ex12-MF-residual")
plt.clf()
draw_MF_graph(G_f)
plt.savefig(f"./ex12-MF-flow-result")
plt.clf()
```

```python
from copy import deepcopy
import numpy as np
from scipy.optimize import linprog
def draw_kwargs_gen(gr, pos):
    return {
        "G": gr,
        "pos": pos,
        "with_labels": True,
        "labels": nx.get_node_attributes(G, "order"),
        "node_color": list(nx.get_node_attributes(gr, "color").values()),
        "edgecolors": list(nx.get_node_attributes(gr, "stroke").values()),
        "width": list(nx.get_edge_attributes(gr, "width").values()),
        "edge_color": list(nx.get_edge_attributes(gr, "color").values()),
        # "connectionstyle": f"arc3, rad = {arc_rad}",
        # "style": list(nx.get_edge_attributes(G, "style").values()),
    }

def draw_graph(gr,pos):
    nx.draw(**draw_kwargs_gen(gr, pos))
    edge_labels = nx.get_edge_attributes(gr, "weight")
    nx.draw_networkx_edge_labels(gr, pos, edge_labels=edge_labels)

def select_edge(gr,positive_edgelist,negative_edgelist):
    edge_weight=[]
    for i in gr.edges():
        if i in positive_edgelist:
            edge_weight.append(1)
        elif i in negative_edgelist:
            edge_weight.append(-1)
        else:
            edge_weight.append(0)
    return edge_weight

G = nx.DiGraph()
v = [i for i in range(12)]
e = [
    (0, 1, {"weight": 3}),
    (0, 2, {"weight": 2}),
    (3, 1, {"weight": 6}),
    (3, 4, {"weight": 2}),
    (1, 4, {"weight": 6}),
    (1, 5, {"weight": 2}),
    (2, 5, {"weight": 1}),
    (9, 6, {"weight": 4}),
    (4, 5, {"weight": 3}),
    (4, 7, {"weight": 3}),
    (5, 7, {"weight": 2}),
    (10, 6, {"weight": 5}),
    (8, 2, {"weight": 9}),
    (8, 9, {"weight": 3}),
    (10, 11, {"weight": 5}),
    (5, 11, {"weight": 4}),
    (2, 9, {"weight": 5}),
    (2, 10, {"weight": 4}),
    (10, 9, {"weight": 4}),
    (10, 5, {"weight": 4}),
    (2, 1, {"weight": 4}),
]
pos=[
    [1.5,2.5],
    [1,2],
    [2,2],
    [0.5,1.5],
    [1,1],
    [2,1],
    [3.5,1.5],
    [1.5,0.5],
    [2.5,2.5],
    [3,2],
    [3,1],
    [2.5,0.5],
]
x_mid,x_rad=2,1.5
y_mid,y_rad=1.5,1
pos_fin={v:[(i[0]-x_mid)/x_rad,(i[1]-y_mid)/y_rad]for v,i in zip(v,pos)}
# print(pos_fin)
G.add_nodes_from(v, **node_type[1])
for i in e:
    G.add_edge(i[0], i[1], **i[2], residue=False, **edge_type[1])

edge_order=G.edges()
N_edge=len(edge_order)
source_dest_pairs=[(0,7),(3,6),(8,11)]
source_dest_nodes=[j for i in source_dest_pairs for j in i]
remaining_nodes=[i for i in G.nodes() if i not in source_dest_nodes]



for idx,i in enumerate(source_dest_pairs,start=2):
    G.add_node(i[0],order="S",**node_type[idx])
    G.add_node(i[1],order="T",**node_type[idx])
draw_graph(G,pos_fin)
plt.savefig("./ex12-MC-start")
plt.clf()
#目的関数z
s=f"\\[z="

tmp_edgelist=[]
for i in source_dest_pairs:
    for j in G.out_edges(i[0]):
        s+=f"x_{{{j[0]}-{j[1]}}}+"
        tmp_edgelist.append(j)
s=s[:-1]
s+=f"\\]\n"
z=select_edge(G,[],tmp_edgelist) #we want to maximize flow so we want to minimize -flow

#store all equality condition here
A_eq=[]




#制約条件source and sink

for i,j in source_dest_pairs:
    s+=f"\\["
    for k in G.out_edges(i):
        s+=f"x_{{{k[0]}-{k[1]}}}+"
    s=s[:-1]
    s+=f"="
    for k in G.in_edges(j):
        s+=f"x_{{{k[0]}-{k[1]}}}+"
    s=s[:-1]
    s+=f"\\]\n"
    A_eq.append(select_edge(G,G.out_edges(i),G.in_edges(j)))
    

    

#制約条件 other nodes
for i in remaining_nodes:
    s+=f"\\["
    for k in G.out_edges(i):
        s+=f"x_{{{k[0]}-{k[1]}}}+"
    s=s[:-1]
    s+=f"="
    for k in G.in_edges(i):
        s+=f"x_{{{k[0]}-{k[1]}}}+"
    s=s[:-1]
    s+=f"\\]\n"
    A_eq.append(select_edge(G,G.out_edges(i),G.in_edges(i)))
b_eq=[0 for i in A_eq]

#制約条件 容量
bounds=[]
for i,j in edge_order:
    w=G.get_edge_data(i,j)["weight"]
    s+=f"\\[0\\leq x_{{{i}-{j}}}\\leq{w}\\]\n"
    bounds.append((0,w))

# print(select_edge(G,[(0,1)]))
res=linprog(z,A_eq=A_eq,b_eq=b_eq,bounds=bounds)
# print(res)
flows=[round(i,2) for i in res.x]
edge_flow={i:j for i,j in zip(edge_order,flows)}


for i,j in zip(edge_order,flows):
    G.add_edge(*i,weight=j)

print(res)

rmv_lst = []
for i in G.edges:
    if G.get_edge_data(*i)["weight"] == 0:
        rmv_lst.append(i)
for i in rmv_lst:
    G.remove_edge(*i)
draw_graph(G,pos_fin)
plt.savefig("./ex12-MC-result")
plt.clf()
```
