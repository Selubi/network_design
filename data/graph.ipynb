```python
import networkx as nx
from matplotlib import pyplot as plt
```

```python
node_type={1: {'color': 'w', 'stroke': 'black'}, 2: {'color': 'r', 'stroke': 'black'}}
```

```python
#node_type={1: {'color': 'w', 'stroke': 'black'}, 2: {'color': 'r', 'stroke': 'black'}}
G=nx.Graph()
v=[i for i in range(7)]
G.add_nodes_from(v,**node_type[1])
e=[(0,1,1),(2,1,1),(2,3,1),(4,5,1),(3,0,1),(1,5,1),(6,1,1),(6,4,1),(3,5,1),(6,2,1)]
G.add_weighted_edges_from(e)

#draw
draw_kwargs={
    'G':G,
    'pos':nx.circular_layout(G),
    'with_labels':True,
    'node_color':list(nx.get_node_attributes(G,'color').values()),
    'edgecolors':list(nx.get_node_attributes(G,'stroke').values())
}
nx.draw(**draw_kwargs)
# nx.draw_networkx_nodes(G,node_color="w",edgecolors="b")
plt.savefig("./1.png")
```
